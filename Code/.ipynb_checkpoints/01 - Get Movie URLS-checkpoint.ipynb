```python
links = pd.read_csv('../Resources/Data/Raw/ml-latest/links.csv')
```

```python
movies_df = pd.read_table('../Resources/Data/Raw/hetrec2011-movielens-2k-v2/movies.dat')

movies_df.rename(columns={'imdbID': 'imdbId'}, inplace=True)
movies_df.rename(columns={'id': 'movieId'}, inplace=True)

# Saving a smaller version that only includes the MovieID and IMDB URL
links_redux = movies_df[["movieId","imdbPictureURL"]]
links_redux.to_csv("../Resources/Data/Derived/links_redux.csv")
```

```python
movies_df.head()
```

```python
links.head()
```

```python
combined_links = pd.merge(links,movies_df[["imdbId","title","imdbPictureURL","movieId"]],on="movieId")
combined_links.drop("imdbId_y",inplace=True, axis=1)
combined_links.rename(columns={"imdbId_x":"imdbId"}, inplace=True)
```

```python
combined_links.head()
```

```python
missing_links = combined_links[combined_links["imdbPictureURL"].isnull() == 1]
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
list_of_indicies = missing_links.index
```

```python
picture_url_list = []

#for i in range(10):
for i in range(len(missing_links)):
    if type(missing_links["imdbPictureURL"][list_of_indicies[1]]) != float:
        print "Data Had, ignoring"
    
    imdb_id = int(list(missing_links["imdbId"])[i])
#    print imdb_id   
    movie_object = ia.get_movie(imdb_id)        
    print movie_object
    try:
        picture_url = movie_object["cover url"]
        if picture_url[0:4] != "http":
            print "bad url"
            picture_url = None
    except:
        picture_url = None
        
    if i % 50 == 0: # Just to be able to see progress
        print i,
    missing_links["imdbPictureURL"][i] = picture_url
    
    picture_url_list.append(picture_url)
print "complete"
```

```python
missing_links["imdbPictureURL"] = picture_url_list
```

```python
pre_existing = combined_links[combined_links["imdbPictureURL"].isnull() != 1].copy()  # Take out those from combined that have the content already
combined_links_filled_in = pre_existing.append(missing_links)
```
